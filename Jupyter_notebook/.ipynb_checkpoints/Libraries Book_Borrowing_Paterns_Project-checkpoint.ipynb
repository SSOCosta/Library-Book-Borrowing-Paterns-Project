# CLEANING AND WRANGLING DATA

```python
# After some tranlation and some cleaning in the excel file, and getting data to Mysql workbench:
import pandas as pd
import sqlalchemy
from sqlalchemy import create_engine
import getpass

# Get MySQL password securely
password = getpass.getpass()

# Establishing connection and creating engine with MySQL
connection_string = 'mysql+pymysql://root:' + password + '@localhost/libraries'
engine = create_engine(connection_string)

# Define the SQL query with proper spacing
query = 'SELECT * FROM libraries.nl_data_full;'

# Read data into a pandas DataFrame
df = pd.read_sql_query(query, engine)
df
```

```python
#Filtering data according to project question
query1 = '''SELECT*
FROM libraries.collection as c
JOIN libraries.loans USING (year)
JOIN libraries.registration USING (year);'''
data = pd.read_sql_query(query1, engine)
data
```

```python
# Checking the shape
data.shape
```

```python
# Replacing columns names/values
data.columns = data.columns.str.replace(' ', '_').str.lower()    
data
```

```python
# Replacing empty spaces by NAN
data.replace('', np.nan, inplace=True)
data
```

```python
# Checking for duplicates
duplicates = data.duplicated().any()
print(duplicates)
# There are no duplicates in this dataset, and for that reason don't need to drop any row
```

```python
# Checking for null values
null_counts = data.isnull().sum()
for column, count in null_counts.items():
    print(f"Column: {column}, Null Count: {count}")
```

```python
# Retrieving just the columns with NAN values on it
columns_with_nan = data.columns[data.isnull().any()]
data_with_nan = data[columns_with_nan]
data_with_nan
```

```python
#Dealing with the nan values in adult_reference_books columns. Accordingly to their explanation, they existed since 2008 that they were 
# merged, into adult_book_non_fiction. For that reason I replaced them with the correspondent proportion
'''Reference works
With effect from 2008, this category has been merged with the category 'Books for adults; Non-fiction'.'''

# Calculating the proportion : adult_reference_books/adult_book_non_fiction(0.02)

# Atrributing the proportion value to fill nan values in adult_reference_books
from math import floor
fill_values = (0.02 * data['adult_book_non_fiction']).astype(int)

# Fill NaN values in column2 with the calculated values
data['adult_reference_books'].fillna(fill_values, inplace=True)
data
```

```python
# The same apllies to fill the nan values in infant_reference_books. Accordingly to their explanation, they existed since 2008 that they were merged, 
# into infant_book_non_fiction. For that reason I replaced them with the correspondent proportion

'''With effect from 2008, this category has been merged with the category 'Books for young people; Development books. '''

# Calculating the proportion : infants_reference_books/infants_book_non_fiction(0.0004)

# Atrributing the proportion value to fill nan values in adult_reference_books
fill_values = (0.0004 * data['infant_books_non_fiction']).astype(int)

# Fill NaN values in column2 with the calculated values
data['infant_reference_books'].fillna(fill_values, inplace=True)
data
```

```python
# The same apllies to fill the nan values in CD-ROMs, CD-I's. Accordingly to their explanation, they existed since 2008 that they were merged, 
# into other_audiovisual. For that reason I replaced them with the correspondent proportion

'''The category 'Audiovisual materials; CD-ROMs, CD-Is' has been merged with the category 
'Audiovisual materials; Other audiovisual materials'. '''

# Calculating the proportion : cd_roms_and_cd_is/other_audiovisual(1.7)

# Atrributing the proportion value to fill nan values in adult_reference_books
fill_values = (1.7 * data['other_audiovisual']).astype(int)

# Fill NaN values in column2 with the calculated values
data['cd_roms_and_cd_is'].fillna(fill_values, inplace=True)
data
```

```python
# Retrieving just the columns with NAN values on it
columns_with_nan = data.columns[data.isnull().any()]
data_with_nan = data[columns_with_nan]
data_with_nan
```

```python
# The same apllies to fill the nan values in loans_adult_reference_books. Accordingly to their explanation, they existed 
# but since 2008 that they were merged,  into loans_adult_books_non_fiction. For that reason I replaced them with the correspondent proportion

'''With effect from 2008, this category has been merged with the category 'Books for adults; Non-fiction'.'. '''

# Calculating the proportion : loans_adult_reference_books/loans_adult_books_non_fiction(0.0005)

# Atrributing the proportion value to fill nan values in adult_reference_books
fill_values = (0.0005 * data['loans_adult_books_non_fiction']).astype(int)

# Fill NaN values in column2 with the calculated values
data['loans_adult_reference_books'].fillna(fill_values, inplace=True)
data
```

```python
# The same apllies to fill the nan values in loans_infants_reference_books. Accordingly to their explanation, they existed 
# but since 2008 that they were merged,  into loans_infants_books_non_fiction. For that reason I replaced them with the correspondent proportion

'''With effect from 2008, this category has been merged with the category 'Books for young people; Development books. '''

# Calculating the proportion : loans_infants_reference_books/loans_infants_books_non_fiction(0.0004)

# Atrributing the proportion value to fill nan values in adult_reference_books
fill_values = (0.0004 * data['loans_infants_books_non_fiction']).astype(int)

# Fill NaN values in column2 with the calculated values
data['loans_infants_reference_books'].fillna(fill_values, inplace=True)
data
```

```python
# The same apllies to fill the nan values in loans__cd_roms_and_cd_is. Accordingly to their explanation, they existed since 2008 that they were merged, 
# into loans_other_audiovisual. For that reason I replaced them with the correspondent proportion

'''With effect from 2008, this category has been merged with the category 'Audiovisual materials; Other audiovisual materials. '''

# Calculating the proportion : loans__cd_roms_and_cd_is/loans_other_audiovisual(0,2)

# Atrributing the proportion value to fill nan values in adult_reference_books
fill_values = (0.2 * data['loans_other_audiovisual']).astype(int)

# Fill NaN values in column2 with the calculated values
data['loans__cd_roms_and_cd_is'].fillna(fill_values, inplace=True)
data
```

```python
# Re_checking just the columns with NAN values on it
columns_with_nan = data.columns[data.isnull().any()]
data_with_nan = data[columns_with_nan]
data_with_nan
```

```python
# Dealing with the remaining nan values in loans columns. Accordingly to their explanation, they existed but weren't recorded, 
# for that reason I am going to drop them.
'''Lending
This comprises total recorded physical loans by libraries excluding renewals of physical materials.
Lending of loose magazines and other materials has not been included since 2006.'''

columns_to_drop = ['loose_periodicals','other_collections','loans_loose_periodicals','loans_other_collection']  

data_final = data.drop(columns=columns_to_drop)
data_final
```

```python
# Calculate IQR for key variables
Q1 = data1[['number_public_libraries', 'total_collection', 'total_users']].quantile(0.25)
Q3 = data1[['number_public_libraries', 'total_collection', 'total_users']].quantile(0.75)
IQR = Q3 - Q1

# Identify potential outliers using the IQR method
outliers_iqr = ((data1[['number_public_libraries', 'total_collection', 'total_users']] < (Q1 - 1.5 * IQR)) | (data1[['number_public_libraries', 'total_collection', 'total_users']] > (Q3 + 1.5 * IQR)))

# Print rows with potential outliers
print(data1[outliers_iqr.any(axis=1)])

# In the output, its visible that for the year 2005 and 2006, there are potential outliers based on the specified criteria 
# for number_public_libraries, total_collection, total_users and total loans. These are the rows where at least one of these 
# variables falls outside the calculated bounds.
```
